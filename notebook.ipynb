```python
import numpy as np
a = np.array([1, 2, 3, 4, 5, 6])

# Multiplication
b1 = 10 * a                  # Direct multiplication by 10
b2 = np.multiply(10, a)      # Alternative way using np.multiply

# Addition
c1 = a + b1                  # Direct addition
c2 = np.add(a, b1)           # Alternative way using np.add

# Subtraction
sub1 = a - 2                 # Direct subtraction of 2 from each element
sub2 = np.subtract(a, 2)     # Alternative way using np.subtract

# Division
div1 = a / 2                 # Direct division by 2
div2 = np.divide(a, 2)       # Alternative way using np.divide

# Logarithm (natural log)
d = np.log(a)                # Logarithm for scaling or normalization

# Square root
e = np.sqrt(a)               # Square root of each element

# Squaring each element
f = a ** 2                   # Squaring using exponent
g = np.square(a)             # Alternative using np.square

# Printing results
print('Array a:', a)
print('Multiplication results b1:', b1)
print('Multiplication results b2:', b2)
print('Addition results c1:', c1)
print('Addition results c2:', c2)
print('Subtraction results sub1:', sub1)
print('Subtraction results sub2:', sub2)
print('Division results div1:', div1)
print('Division results div2:', div2)
print('Logarithm result d:', d)
print('Square root result e:', e)
print('Squared result f:', f)
print('Squared result using np.square:', g)

```

## Convert the list of lists to a NumPy array

```python
import numpy as np
import csv

# Open and read the CSV file with the correct path
csv_file_object = csv.reader(open('extended_student_exam_prediction_dataset.csv', newline=''), delimiter=',')

# Initialize an empty list to store the data
data = []

# Read the header row (to skip it) and store it if needed
header = next(csv_file_object)

# Loop through each row in the CSV file and append to the data list
for row in csv_file_object:  
    data.append(row)  # Add each row to the data list

# Convert the list of lists to a NumPy array
data = np.array(data)

# Print the NumPy array
print("Data as NumPy array:\n", data)
```

## Convert the first five columns to float for analysis

```python
import numpy as np
import csv


# Separate the first five columns (numerical data) and the last column (label)
study_data = data[:, 0:5].astype(np.float_)  # Convert the first five columns to float for analysis
pass_label = data[:, 5]                      # Select only the last column for label

# Print the separated arrays for verification
print("Study Data (first five columns):\n", study_data)
print("\nPass Label (last column):\n", pass_label)


# Sum up the values of the first column (Study Hours)
sum_study_hours = study_data[:, 0].sum()
print("\nSum of Study Hours:", sum_study_hours)

# Extract the first column (Study Hours) and assign it to a new variable
study_hours = study_data[:, 0]

# Display the extracted column
print("\nStudy Hours (first column):\n", study_hours)
```

## Basic Statistics (Mean, Sum, Standard Deviation)

```python
# Mean of each feature
mean_study_hours = study_data[:, 0].mean()
mean_previous_grades = study_data[:, 1].mean()

print("Mean Study Hours:", mean_study_hours)
print("Mean Previous Grades:", mean_previous_grades)

# Standard deviation of Attendance Rate
std_attendance = study_data[:, 2].std()
print("Standard Deviation of Attendance Rate:", std_attendance)
```

## Conditional Selection

- Use conditions to filter data. For example, find students who studied more than 8 hours or had high attendance rates.
- This is useful for identifying students who meet certain performance thresholds.

```python
# Select students who studied more than 8 hours
high_study_hours = study_data[study_data[:, 0] > 8]

# Select students with Attendance Rate above 85
high_attendance = study_data[study_data[:, 2] > 85]

print("Students with high study hours:\n", high_study_hours)
print("Students with high attendance rates:\n", high_attendance)
```

## Normalization and Standardization

- Normalization: Rescale each feature to a range between 0 and 1. Useful when features have different ranges, like Study Hours vs. Attendance Rate.
- Standardization: Scale features to have a mean of 0 and standard deviation of 1. Useful for preparing data for machine learning.

```python
# Normalization (scales values between 0 and 1)
normalized_study_hours = study_data[:, 0] / study_data[:, 0].max()
normalized_previous_grades = study_data[:, 1] / study_data[:, 1].max()

print("Normalized Study Hours:\n", normalized_study_hours)
print("Normalized Previous Grades:\n", normalized_previous_grades)

# Standardization (scales values to have a mean of 0 and standard deviation of 1)
standardized_study_hours = (study_data[:, 0] - study_data[:, 0].mean()) / study_data[:, 0].std()
print("Standardized Study Hours:\n", standardized_study_hours)
```

## Correlation between Features

```python
# Correlation between Study Hours and Previous Grades
correlation = np.corrcoef(study_data[:, 0], study_data[:, 1])[0, 1]
print("Correlation between Study Hours and Previous Grades:", correlation)
```

## Adding New Features (Feature Engineering)

Example: Create a new feature that represents the “engagement level,” combining Study Hours, Attendance Rate, and Extracurriculars. This can help provide a summary score for each student’s involvement.

```python
# Engagement level as a weighted combination of Study Hours, Attendance Rate, and Extracurriculars
engagement_level = 0.5 * study_data[:, 0] + 0.3 * study_data[:, 2] + 0.2 * study_data[:, 4]
print("Engagement Level for each student:\n", engagement_level)
```

## Categorizing Students by Performance with Conditional Slicing

```python
# Define categories based on Study Hours
high_performers = study_data[study_data[:, 0] >= 8]  # Study hours >= 8
average_performers = study_data[(study_data[:, 0] < 8) & (study_data[:, 0] >= 5)]
low_performers = study_data[study_data[:, 0] < 5]

print("High Performers:\n", high_performers)
print("Average Performers:\n", average_performers)
print("Low Performers:\n", low_performers)
```

## Matrix Operations for Advanced Calculations


- For advanced analysis, you could use matrix operations to calculate weighted scores, or perform matrix multiplication to model data interactions.
- Example: Calculate a weighted score matrix where each feature has a different impact on the "Pass" prediction.

```python
# Define weights for each feature (Study Hours, Previous Grades, Attendance Rate, etc.)
weights = np.array([0.4, 0.3, 0.2, 0.05, 0.05])

# Calculate weighted score for each student
weighted_scores = np.dot(study_data, weights)
print("Weighted Scores for each student:\n", weighted_scores)
```

## Unique and Counting Occurrences

- Use np.unique to find unique values and their counts in a column (e.g., number of students participating in extracurriculars).
- This is helpful for categorical analysis.

```python
# Count the number of students in each category of Extracurriculars (0, 1, or 2)
extracurricular_counts = np.unique(study_data[:, 4], return_counts=True)
print("Extracurricular Participation Counts:", extracurricular_counts)
```

## Stacking Arrays to Add New Columns

```python
# Stack engagement_level as a new column in the study_data array
study_data_with_engagement = np.hstack((study_data, engagement_level.reshape(-1, 1)))
print("Study Data with Engagement Level:\n", study_data_with_engagement)
```
